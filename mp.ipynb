### <font color=blue>_Principe:_ </font>
    - le jeu se joue en duel, le score maximal de l'un des joueurs est n 
    - Une fois à trois les joueurs révèlent leur main (pierre, feuille ou ciseaux) en même temps: La plus forte des formes l’emporte et le joueur marque le point gagnant
    - Si les deux joueurs utilisent la même forme c’est un match nul

### <font color=blue>_Les règles du jeu:_ </font>
     - la pierre écrasse les ciseaux et gagne
     - La feuille enveloppe la pierre et gagne
     - les ciseaux découpent la feuille et gagnent

#### <font color=red>_Objectifs:_ </font>
      - Objectif1: Simuler le jeu
      - Objectif2: déterminer la probablité de gagner dans ce jeu

### <font color=red> Simulation du choix : </font>

```python
import numpy as np
```

```python
def choix():
    return int(np.random.choice(range(1,4),1))
#avoir une valeur de l'ensemble{1,2,3} aléatoirement et on transforme le tableau par la fonction int() en un entier
```

```python
choix()
```

```python
def description(m):
        if m==1:
            return "pierre"
        if m==2:
            return "feuille"
        if m==3:
            return "ciseaux"
```

```python
description(2)
```

- 1 désigne la pierre
- 2 désigne la feuille
- 3 désigne les ciseaux


### <font color=red> Simulation du jeux:

```python
def Sim(n):
    j1=0 #choix du joueur1 
    j2=0 #choix du joueur2
    score1=0 #score du joueur1
    score2=0 #score du joueur2
    while score1<=n and score2<=n:
        j1=choix()
        j2=choix()
        print(j1,j2)
        if j1==1:
            if j2==1:
                print('le match est nul')
                print('le score du joueur1:',score1)
                print('le score du joueur2:',score2)
            if j2==2:
                score2=score2+1
                print('le gagnant est',description(j2))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
            if j2==3:
                score1=score1+1
                print('le gagnant est:',description(j1))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
        if j1==2:
            if j2==1:
                score1=score1+1
                print('le gagnant est:',description(j1))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
            if j2==2:
                print('le match est nul')
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
            if j2==3:
                score2=score2+1
                print('le gagnant est:',description(j2))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
        if j1==3:
            if j2==1:
                score2=score2+1
                print('le gagnant est:',description(j2))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
            if j2==2:
                score1=score1+1
                print('le gagnant est:',description(j1))
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
            if j2==3:
                print('le match est nul')
                print('le score du joueur1:',score1)                
                print('le score du joueur2:',score2)
    if score1>score2:
         return "joueur1"
    else:
         return "joueur2"
```

```python
Sim(5)
```

```python
#garder le résultat dans une liste L:
L=[]
L.append(Sim(5))
print(L)
```

### <font color=red> Simulation de $10^2$  scénario : </font>

```python
V=10**2
L=[]
for i in range(V):
    L.append(Sim(5))
print(L)
```

```python
L.count('joueur1') #compter le nombre d'occurence du joueur 1 dans la liste(le nombre de gains du joueur1)"
```

```python
def Freq(L):
    return L.count('joueur1')/len(L)
Freq(L)
```

<b> En jouant $10^2$ fois, on remarque que la probabilité de gagner dans ce jeu est égale à 0.51
